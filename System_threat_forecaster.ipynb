## Dataset Description
The dataset consists of data collected from various systems, containing detailed information about system configurations, security settings, hardware specifications, and software versions. This data is sourced from antivirus reports, which track system properties and detect potential malware infections.

With 100,000 entries and 76 features, the dataset includes categorical, numerical, and boolean variables that describe each machine. Key attributes include **operating system details, security settings, processor specifications, disk and memory capacities, and antivirus configurations**. The target variable indicates whether malware was detected on a system.

## Correlation Heatmap (Numerical Features)

```python
# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Drop constant columns (columns with a single unique value)
numeric_df = numeric_df.loc[:, numeric_df.nunique() > 1]

# Compute correlation
correlation_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(75, 75))  
sns.heatmap(correlation_matrix, cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Set a threshold for high correlation
high_corr_threshold = 0.5  

# Identify highly correlated feature pairs
high_corr_features = correlation_matrix[(correlation_matrix.abs() > high_corr_threshold) & (correlation_matrix != 1.0)]

# Drop NaN values
high_corr_pairs = high_corr_features.stack().reset_index()
high_corr_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']
high_corr_pairs = high_corr_pairs.dropna().sort_values(by='Correlation', ascending=False)

print("Highly Correlated Features (>|0.5|):")
print(high_corr_pairs)
```

```python
# List of highly correlated features
high_corr_features = [
    "OSUILocaleID", "OSInstallLanguageID", "OSBuildNumberOnly", "OSBuildNumber",
    "PrimaryDisplayResolutionVertical", "PrimaryDisplayResolutionHorizontal",
    "ProcessorManufacturerID", "ProcessorModelID", "PrimaryDiskCapacityMB", "SystemVolumeCapacityMB",
    "GeoRegionID", "CountryID", "ProcessorCoreCount", "TotalPhysicalRAMMB",
    "PrimaryDisplayDiagonalInches", "InternalBatteryNumberOfCharges", "OSBuildRevisionOnly",
    "NumAntivirusProductsInstalled", "AntivirusConfigID", "IsPassiveModeEnabled",
    "RealTimeProtectionState"
]

# Filter correlation matrix to include only selected features
filtered_corr_matrix = df[high_corr_features].corr()

# Plot heatmap
plt.figure(figsize=(12, 8))  # Adjust figure size
sns.heatmap(filtered_corr_matrix, cmap="coolwarm", annot=True, fmt=".2f", linewidths=0.5)

# Title
plt.title("Heatmap of Highly Correlated Features (> |0.5|)")

# Show plot
plt.show()
```

#### Feature Engineering and data preprocessing pipeline

```python
from sklearn.impute import SimpleImputer

# Identify numerical and categorical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns
num_cols = num_cols.drop('target', errors='ignore')  # Ensure 'target' is not included

# Impute numerical columns with mean
imputer_num = SimpleImputer(strategy='mean')
df[num_cols] = imputer_num.fit_transform(df[num_cols])
test_data[num_cols] = imputer_num.transform(test_data[num_cols])

# Impute categorical columns with most frequent value
imputer_cat = SimpleImputer(strategy='most_frequent')
df[cat_cols] = imputer_cat.fit_transform(df[cat_cols])
test_data[cat_cols] = imputer_cat.transform(test_data[cat_cols])

# Convert date columns to datetime and extract month
df['DateAS'] = pd.to_datetime(df['DateAS']).dt.month
df['DateOS'] = pd.to_datetime(df['DateOS']).dt.month
test_data['DateAS'] = pd.to_datetime(test_data['DateAS']).dt.month
test_data['DateOS'] = pd.to_datetime(test_data['DateOS']).dt.month
```

#### üí°INSIGHTS:
    Imputing missing values: Fills missing numerical data with the mean and categorical data with the most frequent value.
    Date handling: Extracts the month from date columns.

```python
# Calculate the number of unique values per column
cardinality = df.nunique().sort_values(ascending=False)

# Identify columns with only one unique value (low variance)
low_cardinality_cols = cardinality[cardinality == 1].index.tolist()

# Drop these columns from both training and test datasets
df.drop(columns=low_cardinality_cols, inplace=True)
test_data.drop(columns=low_cardinality_cols, inplace=True)

# Print the new shape of df after dropping low-cardinality columns
print("Updated df shape:", df.shape)
```

#### üí°INSIGHTS:
               Columns with only one unique value provide no useful information for the model.
               These columns do not help in distinguishing classes or making predictions.

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Separate categorical features based on cardinality
selected_features = [col for col in categorical_cols if df[col].nunique() <= 10]  # Low-cardinality categorical features
high_cardinality_features = [col for col in categorical_cols if df[col].nunique() > 10]  # High-cardinality categorical features

# Apply One-Hot Encoding only on low-cardinality categorical features
df = pd.get_dummies(df, columns=selected_features, drop_first=True)
test_data = pd.get_dummies(test_data, columns=selected_features, drop_first=True)

# Print updated df shape
print("Updated df shape:", df.shape)
df.head()
```

 #### üí°INSIGHTS:
    The categorical columns are identified  and one-hot encoding used for those with few unique values. 

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for col in high_cardinality_features:
# Apply Label Encoding
    df[col] = encoder.fit_transform(df[col])
    test_data[col] = encoder.fit_transform(test_data[col]) 

# Display the transformed dataset
df.head()
```

#### üí°INSIGHTS:
    This code converts categorical data with many unique values into a numerical format using Label Encoding. 

```python
from sklearn.preprocessing import MinMaxScaler

# Identify numerical columns (excluding target)
num_cols = df.select_dtypes(include=['int64','float64']).columns.drop('target', errors='ignore')

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Fit on training data and transform both datasets
df[num_cols] = scaler.fit_transform(df[num_cols])  # Fit & transform training data
test_data[num_cols] = scaler.transform(test_data[num_cols])  # Transform test data
```

#### üí°INSIGHTS:
    x_scaled = (x - min(x)) / (max(x) - min(x))
    This code scales numerical data to a range of 0 to 1 using Min-Max scaling. 

```python
# Get the intersection of columns in df and test_data while keeping 'target'
common_columns = list(set(test_data.columns) & set(df.columns)) + ['target']

# Select only the common columns
df = df[common_columns]

print("Columns aligned successfully!")
```

#### üí°INSIGHTS:
    The code makes sure the training and test datasets have the same columns
